```python
def check_exists_by_xpath(xpath):
    try:
        WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,xpath)))
    except Exception:
        return False
    return True

def get_comments():
        while check_exists_by_xpath("//div/ul/li/div/button"):
            load_more_comments_element = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,"//div/ul/li/div/button")))
            load_more_comments_element.click()
#             sleep(1)

#         sleep(2)
        soup = BeautifulSoup(driver.page_source,'lxml')
        comms = soup.find_all('div',attrs={'class':'C4VMK'})
        soup_2 = BeautifulSoup(str(comms),'lxml')                            
        spans = soup_2.find_all('span')
        comments = [i.text.strip() for i in spans if i != '']
        return  comments
all_comments=get_comments()
```

```python
all_comments=list(filter(("Verified").__ne__, all_comments))
```

```python
users=all_comments[0::2]
comments=all_comments[1::2]
```
